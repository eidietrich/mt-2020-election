```python
cols = ['Candidate','Contribution Type', 'First Name','Last Name','Date Paid','Amount','Amount Type']
gg[
    (gg['Contribution Type'].isin([9]))
    & (gg['Last Name'] == 'Gianforte')
    & (gg['First Name'].isin(['Greg', 'Susan']))
][cols]['Amount'].sum()
```

```python
ggmt = gg[gg['State'] == 'MT']
ggmt_indv = gg[(gg['State'] == 'MT') & (gg['Contribution Type'] == 9)]
print(len(ggmt))
print(len(ggmt_indv))
print(ggmt['Amount'].sum())
print(ggmt_indv['Amount'].sum())
```

```python
ggmt_indv['Amount'].sum() / len(ggmt_indv)
```

```python
# ggmt['Amount'].sum() / len(ggmt)
gg_indv = gg[gg['Contribution Type'] == 9]
gg_indv['Amount'].sum() / len(gg_indv)
```

```python
gg_indv  = gg[gg['Contribution Type'] == 9]
gg_bystate = gg_indv.groupby('State').agg(donations=('Amount',sum),count=('Amount','count'))
gg_bystate['percent'] = gg_bystate['donations'] / gg_bystate['donations'].sum()
gg_bystate['average_size'] = gg_bystate['donations'] / gg_bystate['count']
gg_bystate.sort_values('count',ascending=False).head(10)
```
